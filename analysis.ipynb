```python
import pandas as pd
```

```python
## Azure
vm_table_data_link = "https://azurepublicdatasettraces.blob.core.windows.net/azurepublicdatasetv2/trace_data/vmtable/vmtable.csv.gz"
vm_table_data = pd.read_csv(vm_table_data_link, compression = "gzip")
vm_table_data.head()
```

```python
vm_table_column_names_schema = ["vm_id", "subscription_id", "deployment_id", "vm_creation_timestamp", "vm_deletion_timestamp", "max_cpu", "avg_cpu", "p95_max_cpu", "vm_category", "vm_core_count", "vm_memory"]
vm_table_data.columns = vm_table_column_names_schema
vm_table_data.head()
```

```python
def get_vm_cpu_readings(index):
    vm_cpu_reading_link = "https://azurepublicdatasettraces.blob.core.windows.net/azurepublicdatasetv2/trace_data/vm_cpu_readings/vm_cpu_readings-file-" + str(index) + "-of-195.csv.gz"
    curr_vm_cpu_reading_data = pd.read_csv(vm_cpu_reading_link, compression = "gzip")
    return curr_vm_cpu_reading_data
```

```python
cpu_vm_reading_data_1 = get_vm_cpu_readings(1)
cpu_vm_reading_data_1.head()
```

```python
cpu_vm_column_names = ["timestamp", "vm_id", "min_cpu", "max_cpu", "avg_cpu"]
cpu_vm_reading_data_1.columns = cpu_vm_column_names
cpu_vm_reading_data_1.head()
```
